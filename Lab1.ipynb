```python
# @markdown Importing Packages and Libraries

# @markdown Setting values

import numpy as np
n=100
d=3
h = 0.001
np.random.seed(42)
X = np.random.randn(n,d)
W = np.random.randn(d,1)
B = np.random.randn(n,1)
Y = np.random.randn(n,1)
```

```python
# @markdown Loss Function

def Loss (X,W,B,Y):
  P = X@W
  Q = P+B-Y
  L = (Q.T@Q)/2
  return L
```

```python
# @markdown Derivative of Loss function with respect to the given input variable using First Principal Rule

def delLX (L,X,n,d):
  delOp = np.empty((n,d))
  for i in range(n):
    for j in range(d):
      X_h = np.copy(X)
      X_h[i,j]  = X_h[i,j] + h
      L_h = Loss(X_h,W,B,Y)
      delOp[i,j] = (L_h - L)/h
  print(f"The derivative 'del L/del X' matrix from first principal is: {delOp}")
```

```python
# @markdown Derivative of Loss function with respect to the given weights using First Principal Rule

def delLW (L,W,d):
  delOp = np.empty(d)
  for i in range(d):
    W_h = np.copy(W)
    W_h[i]  += h
    L_h = Loss(X,W_h,B,Y)
    L = Loss(X,W,B,Y)
    delOp[i] = (L_h - L)/h
  print(f"The derivative 'del L/del W' vector from first principal is: {delOp}")
```

```python
# @markdown Derivative of Loss function with respect to the given bias using First Principal Rule

def delLB (L,B,n):
  delOp = np.empty(n)
  for i in range(n):
    B_h = np.copy(B)
    B_h[i]  += h
    L_h = Loss(X,W,B_h,Y)
    L = Loss(X,W,B,Y)
    delOp[i] = (L_h - L)/h
  print(f"The derivative 'del L/del B' vector from first principal is: {delOp}")
```

```python
# @markdown Derivative of Loss function with respect to the given input variable using Formula

def printDelLX ():
  delLX_f =((X@W) +B - Y)@W.T
  print(f"The derivative 'del L/del W' vector from formula is: {delLX_f}")
```

```python
# @markdown Derivative of Loss function with respect to the given weights using Formula

def printDelLW ():
  delLW_f = X.T@((X@W) + B - Y)
  print(f"The derivative 'del L/del W' vector from formula is: {delLW_f}")
```

```python
# @markdown Derivative of Loss function with respect to the given bias using Formula

def printDelLB ():
  delLB_f = ((X@W) +B  - Y)
  print(f"The derivative 'del L/del B' vector from formula is: {delLB_f}")
```

```python
# @markdown Printing the derivatives or the Jacobian of the Loss Function with respect to given input variable, weights and bias

L = Loss(X,W,B,Y)
delLX(L,X,n,d)
printDelLX()
delLW(L,W,d)
printDelLW()
delLB(L,B,n)
printDelLB()
```
